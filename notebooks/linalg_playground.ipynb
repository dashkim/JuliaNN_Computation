# Linear algebra playground (Julia)

Scratch space for matrix computations that show up in NNs: products, norms, conditioning, spectra. Extend with conjugate gradient, kernels, or structured operators as you go.


```julia
using LinearAlgebra, Random

Random.seed!(0)
A = randn(5, 5)
S = A * A' + 5I   # symmetric positive definite
b = randn(5)
```

## Solve, norm, condition number


```julia
x = S \ b
@show norm(S * x - b)
@show cond(S)
```

## Eigenvalues (symmetric `S`)


```julia
λ = eigvals(S)
@show minimum(λ), maximum(λ)
```
